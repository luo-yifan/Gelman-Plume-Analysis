```python
import numpy as np
import pathlib
import os

import pandas as pd
import datetime
import matplotlib.pyplot as plt

from autots import AutoTS
```

```python
DIR_NAME = '../result/result(well_rec2)/simple/'
name_list = [name for name in os.listdir("../data/Well_Rec_txt2/")]
print(name_list)
count  = 0
nearby_r_depth = {
    'TW-19' : ['MW-79d', 'MW-85'],
    # '110-Parkland-Plaza': ['4141-Jackson Rd'],
    'A2-Cleaning-Supply':['MW-53i'],
    'MW-92':['MW-130i'],
    # 'MW-122s':[]
}
for i in name_list:
    count+=1
    # print(i)
    name = i.split('.')[1] + '@' + i.split('.')[2]
    if i.split('.')[1] in nearby_r_depth and i.split('.')[2] == 'S2020':
        print(name)
        if os.path.exists(DIR_NAME + '/cache/' + name):
            continue
        well_data = pd.read_csv('../data/Well_Rec_txt2/' + i)
        # print(well_data)
        well_data['Date'] = well_data.apply(lambda row: datetime.datetime(int(row.Year), int(row.Month), 1), axis=1)
        well_data['ds'] = pd.to_datetime(well_data['Date'])
        well_data['y'] = well_data['Concentration']
        well_data.set_index('ds', inplace=True)
        fig, ax = plt.subplots()

        mod = AutoTS(forecast_length=5,
                 frequency='Y',
                 ensemble='simple',
                 no_negatives=True,
                 min_allowed_train_percent=0.2,
                 verbose=-4)
        mod = mod.fit(well_data, date_col='Date', value_col='Concentration', id_col=None)
        prediction = mod.predict()
        forecast = prediction.forecast

        ft = open(DIR_NAME + "mod/" + name + '.txt', "w")
        ft.write(str(mod))
        ft.close()

        # ax.plot(group.y,  label='a')
        # ax.plot(forecast,  label='b')

        ax.scatter(well_data.y.index, well_data.y, label='a')

        ax.scatter(forecast.index, forecast, label='b')

        ax.set_title(name)
        plt.savefig(DIR_NAME + '/fig/' + name + '.png')

        cache_dir = pathlib.Path(DIR_NAME + '/cache/' + name)
        cache_dir.mkdir(parents=True, exist_ok=True)

        well_data.y.to_csv(cache_dir.joinpath('data.csv'))
        forecast.index.name = "Date"
        forecast.squeeze().to_csv(cache_dir.joinpath('predict.csv'))

        print("[predict success]", name)

    # except KeyboardInterrupt:
    #     break
    #
    # except:
    #     print("[predict error]", name)
    #     count+=1
    #     continue
```
