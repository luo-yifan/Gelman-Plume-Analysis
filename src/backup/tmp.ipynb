```python
import pathlib

import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
gelman_data = pd.read_excel("../../result/history/result(new)/joined_well_data.xlsx")
gelman_data.head(5)
```

```python
gelman_data.columns
```

```python
well_name_list = list(dict.fromkeys(gelman_data['Bore'].tolist()))
well_name_list[0:10]
```

```python
res = {}
for well_name in well_name_list:
    var = gelman_data.loc[gelman_data['Bore'] == well_name]
    years_max = {}
    for y in range(1986, 2020):
        for index, row in var.iterrows():
            if row['SampleDate'].year == y:
                if years_max.get(y) is not None:
                    if years_max[y] < row['Value']:
                        years_max[y] = row['Value']
                else:
                    years_max[y] = row['Value']
    res[well_name] = years_max
print(res)
```

```python
for key in res['110 Parkland Plaza']:
    print(key, res['110 Parkland Plaza'][key])

```

```python
# well_name_list = ['1024 Westview Way']
df = pd.DataFrame(res, columns=['Bore', 'Depth1', 'Depth2', 'LAT', 'LONG', '1986',
                                '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
                                '1996', '1997', '1998', '1999', '2000', '2001',
                                '2002', '2003', '2004','2005','2006','2007','2008', '2009', '2010','2011',
                                '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020'])
for well_name in well_name_list:
    var = gelman_data.loc[gelman_data['Bore'] == well_name].iloc[0]
    dict_t = {'Bore': var['Bore'], 'Depth1': var['Depth1'], 'Depth2': var['Depth2'], 'LAT': var['lat'],
              'LONG': var['lon']}
    if var['Depth1'] < 0:
        dict_t['Depth1'] = 'NaN'
    if var['Depth2'] < 0:
        dict_t['Depth2'] = 'NaN'
    for i in range(1986,2021):
        if res[var['Bore']].get(i) is not None:
            if res[var['Bore']].get(i) <= 0.5:
                dict_t[str(i)] = 0
            else:
                dict_t[str(i)] = res[var['Bore']].get(i)
        else:
            dict_t[str(i)] = 'NaN'
    df = df.append(dict_t, ignore_index=True)
print(df)
```
